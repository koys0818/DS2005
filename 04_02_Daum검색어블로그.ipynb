```python
import requests
from bs4 import BeautifulSoup
import urllib.parse as urlparser
```

```python
keyword = "코로나"
url = "https://search.daum.net/search?w=blog&nil_search=btn&DA=NTB&enc=utf8&q=" + urlparser.quote(keyword)
url
```

```python
#Daum 에서 특정한 '키워드'로 검색하기
#거기서 '블로그' 항목으로 들어가서
#블로그 제목, 블로그 이름을 뽑아보세요.

#실제로 포스팅 이름과 블로그 이름이 아래와 같이 나와야 함..

# dict 들의 list
# 각 dict 에는 포스팅 제목 과 블로그 제목이 담겨야 함
#[
#   {
#        post_title: 포스팅 제목,  
#        blog_title: 블로그 제목
#   },
#   {},
#   {},
#]
```

```python
response = requests.get(url)
response
```

```python
dom = BeautifulSoup(response.text, "html.parser")
dom
```

```python
item_list = dom.select("div#blogColl .coll_cont ul.list_info li")
len(item_list)
```

```python
# 포스트 제목
for item in item_list:
    print(item.select_one(".f_link_b").text.strip())
```

```python
# 블로그 제목
for item in item_list:
    print(item.select_one(".info .f_nb").text.strip())
```

# 완성코드 : 포스트 제목 + 블로그 제목 

```python
import requests
from bs4 import BeautifulSoup
import urllib.parse as urlparser
keyword = input("검색어 입력")
url = "https://search.daum.net/search?w=blog&nil_search=btn&DA=NTB&enc=utf8&q=" + urlparser.quote(keyword)
response = requests.get(url)
dom = BeautifulSoup(response.text, "html.parser")
item_list = dom.select("div#blogColl .coll_cont ul.list_info li")

result = [
    {
        "post_title" : item.select_one(".f_link_b").text.strip()
        , "blog_title" : item.select_one(".info .f_nb").text.strip()
    }
    for item in item_list
]    

print(result)
    
```
